```python
#find how many matches a team won in a season
seasons = pd.unique(matches['season'])
```

```python
winners_in_each_season = []
for season in seasons:
    matches_in_a_season = matches[matches['season']==season]
    winners_in_each_season.append(pd.value_counts(matches_in_a_season['winner']))
```

```python
win = matches[matches['toss_winner'] == matches['winner']]
win_count = pd.value_counts(win['toss_decision'])
lose = pd.value_counts(matches[
    matches['toss_winner'] != matches['winner']]
                       ['toss_decision'])
win_count
```

```python
labels = ['field_win', 'bat_win', 'field_lose', 'bat_lose']
```

```python
data = []
for value in win_count:
    data.append(value)
for value in lose:
    data.append(value)
```

```python
data
```

```python
plot.pie(data, labels=labels, autopct='%.2f%%', startangle=90, counterclock=False)
```
